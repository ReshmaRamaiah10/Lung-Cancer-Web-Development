```python
#!/usr/bin/env python3
#Query the miRNA database through the browser using a cgi program

import pymysql
import cgi
import cgitb
import json

#the next line is useful for debugging
#it causes errors during execution to be sent back to the browser
cgitb.enable()

#retrieve input data from the web server
form = cgi.FieldStorage() 

#next line is always required as first part of http output
print("Content-type: text/html\n")

if (form):
    #establish the connection on bioed
    connection = pymysql.connect(host='bioed.bu.edu',user='Group_H',password='Group_H', db='Group_H',port = 4253)  
    cursor = connection.cursor()
    #get submitted values
    selector = form.getvalue("selector","")

    if (selector == "SampleType_Chart"):
        cursor.execute("select SampleType, count(*) from pcga_annotation group by SampleType;")
        results = cursor.fetchall() 
        print(json.dumps(results))
    
    elif (selector == "Cohort_Chart"):
        cursor.execute("select Cohort, count(*) from pcga_annotation group by Cohort;")
        results = cursor.fetchall() 
        print(json.dumps(results))
        
    elif (selector == "Sex_Chart"):
        cursor.execute("select Sex, count(*) from pcga_annotation group by Sex;")
        results = cursor.fetchall() 
        print(json.dumps(results))
        
    elif (selector == "Genomic_Smoking_Status_Chart"):
        cursor.execute("select Genomic_Smoking_Status, count(*) from pcga_annotation group by Genomic_Smoking_Status;")
        results = cursor.fetchall() 
        print(json.dumps(results))
        
    elif (selector == "Previous_LC_History_Chart"):
        cursor.execute("select Previous_LC_History, count(*) from pcga_annotation group by Previous_LC_History;")
        results = cursor.fetchall() 
        print(json.dumps(results))
        
    elif (selector == "HighRisk_LC_Job_Chart"):
        cursor.execute("select HighRisk_LC_Job, count(*) from pcga_annotation group by HighRisk_LC_Job;")
        results = cursor.fetchall() 
        print(json.dumps(results))
        
    elif (selector == "Asbestos_Exp_Chart"):
        cursor.execute("select Asbestos_Exp, count(*) from pcga_annotation group by Asbestos_Exp;")
        results = cursor.fetchall() 
        print(json.dumps(results))
        
    elif (selector == "Molecular_Subtype_Chart"):
        cursor.execute("select Molecular_Subtype, count(*) from pcga_annotation group by Molecular_Subtype;")
        results = cursor.fetchall() 
        print(json.dumps(results))
        
    elif (selector == "Dysplasia_Grade_Chart"):
        cursor.execute("select Dysplasia_Grade, count(*) from pcga_annotation group by Dysplasia_Grade;")
        results = cursor.fetchall() 
        print(json.dumps(results))
        
    elif (selector == "TCGA_Subtype_Chart"):
        cursor.execute("select TCGA_Subtype, count(*) from pcga_annotation group by TCGA_Subtype;")
        results = cursor.fetchall() 
        print(json.dumps(results))
        
    elif (selector == "Location_Desc_Chart"):
        cursor.execute("select Location_Desc, count(*) from pcga_annotation group by Location_Desc;")
        results = cursor.fetchall() 
        print(json.dumps(results))
        
    elif (selector == "Age_at_Baseline_Chart"):
        cursor.execute("select Age_at_Baseline from pcga_annotation order by Age_at_Baseline;")
        results = cursor.fetchall() 
        print(json.dumps(results))
        
    cursor.close()
    connection.close()
```

# tcga chart py

```python
#!/usr/bin/env python3
#Query the miRNA database through the browser using a cgi program

import pymysql
import cgi
import cgitb
import json

#the next line is useful for debugging
#it causes errors during execution to be sent back to the browser
cgitb.enable()

#retrieve input data from the web server
form = cgi.FieldStorage() 

#next line is always required as first part of http output
print("Content-type: text/html\n")

if (form):
    #establish the connection on bioed
    connection = pymysql.connect(host='bioed.bu.edu',user='Group_H',password='Group_H', db='Group_H',port = 4253)  
    cursor = connection.cursor()
    #get submitted values
    selector = form.getvalue("selector","")

    if (selector == "sample_type_Chart"):
        cursor.execute("select sample_type, count(*) from TCGA_annotation group by sample_type;")
        results = cursor.fetchall() 
        print(json.dumps(results))
    
    elif (selector == "year_of_diagnosis_Chart"):
        cursor.execute("select year_of_diagnosis, count(*) from TCGA_annotation group by year_of_diagnosis;")
        results = cursor.fetchall() 
        print(json.dumps(results))
        
    elif (selector == "gender_Chart"):
        cursor.execute("select gender, count(*) from TCGA_annotation group by gender;")
        results = cursor.fetchall() 
        print(json.dumps(results))
        
    elif (selector == "prior_malignancy_Chart"):
        cursor.execute("select prior_malignancy, count(*) from TCGA_annotation group by prior_malignancy;")
        results = cursor.fetchall() 
        print(json.dumps(results))
        
    elif (selector == "ajcc_pathologic_stage_Chart"):
        cursor.execute("select ajcc_pathologic_stage, count(*) from TCGA_annotation group by ajcc_pathologic_stage;")
        results = cursor.fetchall() 
        print(json.dumps(results))
        
    elif (selector == "ajcc_staging_system_edition_Chart"):
        cursor.execute("select ajcc_staging_system_edition, count(*) from TCGA_annotation group by ajcc_staging_system_edition;")
        results = cursor.fetchall() 
        print(json.dumps(results))
        
    elif (selector == "tissue_or_organ_of_origin_Chart"):
        cursor.execute("select tissue_or_organ_of_origin, count(*) from TCGA_annotation group by tissue_or_organ_of_origin;")
        results = cursor.fetchall() 
        print(json.dumps(results))
        
    elif (selector == "ajcc_pathologic_t_Chart"):
        cursor.execute("select ajcc_pathologic_t, count(*) from TCGA_annotation group by ajcc_pathologic_t;")
        results = cursor.fetchall() 
        print(json.dumps(results))
        
    elif (selector == "ajcc_pathologic_m_Chart"):
        cursor.execute("select ajcc_pathologic_m, count(*) from TCGA_annotation group by ajcc_pathologic_m;")
        results = cursor.fetchall() 
        print(json.dumps(results))
        
    elif (selector == "ajcc_pathologic_n_Chart"):
        cursor.execute("select ajcc_pathologic_n, count(*) from TCGA_annotation group by ajcc_pathologic_n;")
        results = cursor.fetchall() 
        print(json.dumps(results))
        
    elif (selector == "age_at_index_Chart"):
        cursor.execute("select age_at_index from TCGA_annotation where age_at_index not regexp 'NA' order by age_at_index;")
        results = cursor.fetchall() 
        print(json.dumps(results))
        
    elif (selector == "primary_diagnosis_Chart"):
        cursor.execute("select primary_diagnosis, count(*) from TCGA_annotation group by primary_diagnosis;")
        results = cursor.fetchall() 
        print(json.dumps(results))
        
    cursor.close()
    connection.close()
```
